## 2. Getting the data

```python
import os
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data (housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

We can now call *fetch_housing_data()* it creates a *datasets/housing* directory in our workspace, downloads the tgz file and extracts the *housing.csv* from it to this directory.

```python
fetch_housing_data()
```

Now let's load data using pandas.

```python
import pandas as pd

def load_housing_data (housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

Set aside part of the data and create a test set. 

In order to avoid data snooping before looking at all the data set aside the test set. 

Approximately 20% of the set randomly choosen.

```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

We can use the about function as below.

```python
training_set, test_set = split_train_test(housing, 0.2)
print(len(training_set), "train +", len(test_set), "test")
```

The above method is not good enough, because with every running of the ML, our machine will eventually get to see all the data set and thats what we want to avoid. 

This can be done so by calculating the hash value of the indices and using the last byte to index our data. So, that even when new data enters we get to use only part of that.

```python
import hashlib

def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio
```

```python
def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

Unfortunately, the data does not have an identifier column.

```python
housing_with_id = housing.reset_index()
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

We will use a more permanent form of identicication from longitudes and latitudes.

```python
housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"]
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "id")
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split (housing, test_size = 0.2, random_state = 42)
```

Random sampling as we have been doing above is usually fine if your data is large enough else stratified sampling, i.e. sampling in such a way that the data is representative of the whole population by dividing the dataset into strata, usually ensures better accuracy.

However, it is important to have a sufficient number of instances in your dataset for each stratum i.e. you shouldn't have too many strata and each stratum should be large enough, to prevent baises in strata importance.

```python
housing["income_cat"] = np.ceil(housing["median_income"]/1.5)
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size= 0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

Let's check to see it works!

```python
strat_test_set["income_cat"].value_counts()/ len(strat_test_set)
```

We can now remove the income_cat attribute to bring the dataset back to its original state.

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop ("income_cat", axis=1, inplace=True)
```
